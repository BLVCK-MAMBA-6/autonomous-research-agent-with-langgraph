````python
"""
Autonomous Research Agent with LangGraph
=========================================
A multi-step research agent that searches, scrapes, analyzes, and synthesizes
information into a comprehensive report with proper citations.

Dependencies:
pip install langgraph langchain langchain-google-genai langchain-community tavily-python beautifulsoup4 requests tiktoken duckduckgo-search

Environment Variables:
- GOOGLE_API_KEY (for Gemini models) - REQUIRED
- TAVILY_API_KEY (for Tavily search) - Optional, will fallback to DuckDuckGo
"""

import os
from typing import TypedDict, List, Annotated
import operator
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_core.prompts import ChatPromptTemplate
from langgraph.graph import StateGraph, END
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from tavily import TavilyClient
import requests
from bs4 import BeautifulSoup
import json
````

````python
import os
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI

def get_llm(model="gemini-2.5-flash"):
    """
    Initialize Google Gemini LLM.
    """
    # 1. Try to load the key from Colab Secrets
    try:
        if not os.environ.get("GOOGLE_API_KEY"):
            # This looks for the secret named 'GOOGLE_API_KEY'
            os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')
    except Exception as e:
        print(f"Warning: Could not load secret. Details: {e}")

    # 2. Validate the key exists
    api_key = os.environ.get("GOOGLE_API_KEY")
    if not api_key:
        raise ValueError(
            "GOOGLE_API_KEY not found! \n"
            "1. Click the Key icon () on the left.\n"
            "2. Create a secret named 'GOOGLE_API_KEY'.\n"
            "3. Paste your key and toggle 'Notebook access' ON."
        )

    # 3. Return the configured model
    return ChatGoogleGenerativeAI(
        model=model,
        temperature=0.3,
        google_api_key=api_key,
        convert_system_message_to_human=True
    )
````

````python
# ============================================================================
# STATE DEFINITION
# ============================================================================

class ResearchState(TypedDict):
    """State that persists through the research workflow."""
    topic: str  # Original research topic
    search_queries: List[str]  # Queries generated
    search_results: List[dict]  # Raw search results with URLs
    scraped_content: Annotated[List[dict], operator.add]  # Accumulated scraped content
    analysis: str  # Current analysis
    iteration: int  # Current iteration count
    max_iterations: int  # Maximum iterations allowed
    needs_more_info: bool  # Flag to continue or stop
    final_report: str  # Final synthesized report
    citations: List[dict]  # List of {id: int, url: str, title: str}
````

````python
class SearchTool:
    """Wrapper for Tavily search API with fallback to DuckDuckGo."""

    def __init__(self, use_tavily=True):
        self.use_tavily = use_tavily
        if use_tavily:
            api_key = os.getenv("TAVILY_API_KEY")
            if not api_key:
                print("Warning: TAVILY_API_KEY not found, falling back to DuckDuckGo")
                self.use_tavily = False
            else:
                self.client = TavilyClient(api_key=api_key)

    def search(self, query: str, max_results: int = 5) -> List[dict]:
        """
        Perform web search and return results with URLs.

        Returns:
            List of dicts with 'title', 'url', 'snippet'
        """
        if self.use_tavily:
            try:
                response = self.client.search(
                    query=query,
                    max_results=max_results,
                    search_depth="advanced"
                )
                return [
                    {
                        "title": r.get("title", ""),
                        "url": r.get("url", ""),
                        "snippet": r.get("content", "")
                    }
                    for r in response.get("results", [])
                ]
            except Exception as e:
                print(f"Tavily search failed: {e}, falling back to DuckDuckGo")
                return self._duckduckgo_search(query, max_results)
        else:
            return self._duckduckgo_search(query, max_results)

    def _duckduckgo_search(self, query: str, max_results: int) -> List[dict]:
        """Fallback search using DuckDuckGo HTML scraping."""
        try:
            from duckduckgo_search import DDGS
            results = []
            ddgs = DDGS()
            search_results = ddgs.text(query, max_results=max_results)

            for r in search_results:
                url = r.get("href", r.get("link", ""))
                if url:
                    results.append({
                        "title": r.get("title", "No title"),
                        "url": url,
                        "snippet": r.get("body", r.get("snippet", ""))
                    })

            return results
        except ImportError:
            print("duckduckgo_search not installed. Install with: pip install ddgs")
            return self._google_search_fallback(query, max_results)
        except Exception as e:
            print(f"DuckDuckGo search failed: {e}")
            return self._google_search_fallback(query, max_results)

    def _google_search_fallback(self, query: str, max_results: int) -> List[dict]:
        """Final fallback using Google Custom Search (requires GOOGLE_SEARCH_API_KEY and GOOGLE_CSE_ID)."""
        try:
            # Try Google Custom Search API if available
            api_key = os.getenv("GOOGLE_SEARCH_API_KEY")
            cse_id = os.getenv("GOOGLE_CSE_ID")

            if not api_key or not cse_id:
                print(" No search APIs available. Please set TAVILY_API_KEY, or GOOGLE_SEARCH_API_KEY + GOOGLE_CSE_ID")
                return []

            url = f"https://www.googleapis.com/customsearch/v1?key={api_key}&cx={cse_id}&q={query}&num={max_results}"
            response = requests.get(url, timeout=10)
            data = response.json()

            results = []
            for item in data.get("items", []):
                results.append({
                    "title": item.get("title", ""),
                    "url": item.get("link", ""),
                    "snippet": item.get("snippet", "")
                })
            return results
        except Exception as e:
            print(f"Google Custom Search failed: {e}")
            return []
````

````python
# ============================================================================
# WEB SCRAPER
# ============================================================================

class WebScraper:
    """Robust web scraper with error handling."""

    @staticmethod
    def scrape(url: str, max_chars: int = 10000) -> dict:
        """
        Scrape content from a URL.

        Returns:
            Dict with 'url', 'content', 'success', 'error'
        """
        try:
            # Use WebBaseLoader from LangChain
            loader = WebBaseLoader(url)
            docs = loader.load()

            if docs:
                content = docs[0].page_content[:max_chars]
                return {
                    "url": url,
                    "content": content,
                    "success": True,
                    "error": None
                }
            else:
                return {
                    "url": url,
                    "content": "",
                    "success": False,
                    "error": "No content extracted"
                }

        except Exception as e:
            # Fallback to requests + BeautifulSoup
            try:
                headers = {
                    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
                }
                response = requests.get(url, headers=headers, timeout=10)
                response.raise_for_status()

                soup = BeautifulSoup(response.content, 'html.parser')

                # Remove script and style elements
                for script in soup(["script", "style"]):
                    script.decompose()

                text = soup.get_text()
                lines = (line.strip() for line in text.splitlines())
                chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
                text = ' '.join(chunk for chunk in chunks if chunk)

                return {
                    "url": url,
                    "content": text[:max_chars],
                    "success": True,
                    "error": None
                }

            except Exception as fallback_error:
                return {
                    "url": url,
                    "content": "",
                    "success": False,
                    "error": str(fallback_error)
                }
````

````python
# ============================================================================
# GRAPH NODES (Agent Steps)
# ============================================================================

def generate_search_queries(state: ResearchState) -> ResearchState:
    """
    Generate search queries based on the topic and previous analysis.
    """
    llm = get_llm()  # Using Gemini - you can change model: get_llm("gemini-1.5-pro")

    if state["iteration"] == 0:
        # First iteration - generate initial queries
        prompt = f"""Given the research topic: "{state['topic']}"

Generate 3 specific, diverse search queries that will help research this topic comprehensively.
Each query should target different aspects or subtopics.

Return only the queries as a JSON array, e.g., ["query 1", "query 2", "query 3"]"""
    else:
        # Subsequent iterations - generate targeted queries based on gaps
        prompt = f"""Research topic: "{state['topic']}"

Previous analysis:
{state['analysis']}

Based on the analysis, what information is still missing or needs deeper investigation?
Generate 2-3 new, specific search queries to fill these gaps.

Return only the queries as a JSON array."""

    response = llm.invoke([HumanMessage(content=prompt)])

    try:
        # Extract JSON array from response
        content = response.content
        if "```json" in content:
            content = content.split("```json")[1].split("```")[0]
        elif "```" in content:
            content = content.split("```")[1].split("```")[0]

        queries = json.loads(content.strip())
        state["search_queries"] = queries
    except:
        # Fallback parsing
        state["search_queries"] = [state["topic"]]

    print(f"\n Generated queries: {state['search_queries']}")
    return state


def perform_searches(state: ResearchState) -> ResearchState:
    """
    Execute searches for all queries and collect results.
    """
    search_tool = SearchTool(use_tavily=True)
    all_results = []

    for query in state["search_queries"]:
        print(f"\n Searching: {query}")
        results = search_tool.search(query, max_results=3)
        all_results.extend(results)

    # Deduplicate by URL
    seen_urls = set()
    unique_results = []
    for result in all_results:
        if result["url"] not in seen_urls:
            seen_urls.add(result["url"])
            unique_results.append(result)

    state["search_results"] = unique_results
    print(f" Found {len(unique_results)} unique sources")
    return state


def scrape_content(state: ResearchState) -> ResearchState:
    """
    Scrape content from search result URLs.
    """
    scraper = WebScraper()
    scraped = []

    for idx, result in enumerate(state["search_results"][:10], 1):  # Limit to top 10
        print(f"\n Scraping ({idx}/{min(10, len(state['search_results']))}): {result['url']}")
        scraped_data = scraper.scrape(result["url"])

        if scraped_data["success"]:
            scraped.append({
                "citation_id": len(state.get("citations", [])) + len(scraped) + 1,
                "title": result["title"],
                "url": result["url"],
                "content": scraped_data["content"]
            })

    # Add to accumulated scraped content
    if "scraped_content" not in state:
        state["scraped_content"] = []
    state["scraped_content"].extend(scraped)

    print(f" Successfully scraped {len(scraped)} pages")
    return state


def analyze_content(state: ResearchState) -> ResearchState:
    """
    Analyze scraped content using map-reduce approach for large texts.
    """
    llm = get_llm()  # Using Gemini

    if not state["scraped_content"]:
        state["analysis"] = "No content was successfully scraped."
        return state

    # Map phase: Summarize each document
    summaries = []
    for doc in state["scraped_content"][-10:]:  # Analyze last 10 documents
        prompt = f"""Summarize the key insights from this source relevant to the topic: "{state['topic']}"

Source: {doc['title']}
Content: {doc['content'][:3000]}

Provide a concise summary of the most important information."""

        response = llm.invoke([HumanMessage(content=prompt)])
        summaries.append({
            "citation_id": doc["citation_id"],
            "title": doc["title"],
            "summary": response.content
        })

    # Reduce phase: Synthesize all summaries
    combined_summaries = "\n\n".join([
        f"Source [{s['citation_id']}] - {s['title']}:\n{s['summary']}"
        for s in summaries
    ])

    synthesis_prompt = f"""Research Topic: "{state['topic']}"

Current iteration: {state['iteration'] + 1}/{state['max_iterations']}

Analyze the following summaries from various sources:

{combined_summaries}

Provide:
1. Key findings and insights discovered so far
2. Patterns or themes across sources
3. What critical information or perspectives are still missing (if any)

Be specific and analytical."""

    response = llm.invoke([HumanMessage(content=synthesis_prompt)])
    state["analysis"] = response.content

    print(f"\n Analysis complete")
    return state


def decide_next_step(state: ResearchState) -> ResearchState:
    """
    Decide whether to continue searching or move to final report.
    """
    llm = get_llm()

    state["iteration"] += 1

    if state["iteration"] >= state["max_iterations"]:
        state["needs_more_info"] = False
        print(f"\n Reached max iterations ({state['max_iterations']})")
        return state

    # Ask LLM if more research is needed
    prompt = f"""Research Topic: "{state['topic']}"

Current Analysis:
{state['analysis']}

Number of sources collected: {len(state['scraped_content'])}
Current iteration: {state['iteration']}/{state['max_iterations']}

Based on this analysis, is there critical information still missing that would significantly improve the research report?

Answer with just "YES" or "NO" and a brief reason."""

    response = llm.invoke([HumanMessage(content=prompt)])
    decision = response.content.strip().upper()

    state["needs_more_info"] = decision.startswith("YES")

    if state["needs_more_info"]:
        print(f"\n Continuing research: {response.content}")
    else:
        print(f"\n Research complete: {response.content}")

    return state


def generate_final_report(state: ResearchState) -> ResearchState:
    """
    Generate the final research report with proper citations.
    """
    llm = get_llm()

    # Build citations list
    citations = []
    for doc in state["scraped_content"]:
        citations.append({
            "id": doc["citation_id"],
            "url": doc["url"],
            "title": doc["title"]
        })

    state["citations"] = citations

    # Create context with all content
    context_parts = []
    for doc in state["scraped_content"]:
        context_parts.append(
            f"[{doc['citation_id']}] {doc['title']}\n{doc['content'][:2000]}\n"
        )

    context = "\n\n".join(context_parts)

    # Generate report
    prompt = f"""You are writing a comprehensive research report.

Topic: "{state['topic']}"

Sources (use these citation IDs in your report):
{context}

Instructions:
1. Write an in-depth, well-structured research report (800-1500 words)
2. Use markdown formatting with headers (##, ###)
3. Cite sources using [1], [2], etc. after relevant claims
4. Include multiple citations per paragraph where appropriate
5. Cover all major aspects found in the research
6. Be analytical and synthesize information across sources
7. Do NOT include a references section - that will be added automatically

Write the report now:"""

    response = llm.invoke([HumanMessage(content=prompt)])
    report = response.content

    # Add citations section
    citations_md = "\n\n---\n\n## References\n\n"
    for citation in citations:
        citations_md += f"[{citation['id']}] {citation['title']}  \n{citation['url']}\n\n"

    state["final_report"] = report + citations_md

    print(f"\n Final report generated with {len(citations)} citations")
    return state
````

````python
# ============================================================================
# GRAPH CONSTRUCTION
# ============================================================================

def create_research_graph():
    """Build the LangGraph state machine."""

    workflow = StateGraph(ResearchState)

    # Add nodes
    workflow.add_node("generate_queries", generate_search_queries)
    workflow.add_node("search", perform_searches)
    workflow.add_node("scrape", scrape_content)
    workflow.add_node("analyze", analyze_content)
    workflow.add_node("decide", decide_next_step)
    workflow.add_node("final_report", generate_final_report)

    # Define edges
    workflow.set_entry_point("generate_queries")
    workflow.add_edge("generate_queries", "search")
    workflow.add_edge("search", "scrape")
    workflow.add_edge("scrape", "analyze")
    workflow.add_edge("analyze", "decide")

    # Conditional edge: loop back or finish
    def should_continue(state: ResearchState) -> str:
        if state["needs_more_info"]:
            return "generate_queries"
        else:
            return "final_report"

    workflow.add_conditional_edges(
        "decide",
        should_continue,
        {
            "generate_queries": "generate_queries",
            "final_report": "final_report"
        }
    )

    workflow.add_edge("final_report", END)

    return workflow.compile()


# ============================================================================
# MAIN EXECUTION
# ============================================================================

def run_research_agent(topic: str, max_iterations: int = 3):
    """
    Execute the research agent.

    Args:
        topic: Research topic/question
        max_iterations: Maximum search-analyze cycles

    Returns:
        Final report as markdown string
    """
    print(f"\n{'='*60}")
    print(f" AUTONOMOUS RESEARCH AGENT")
    print(f"{'='*60}")
    print(f"Topic: {topic}")
    print(f"Max iterations: {max_iterations}")
    print(f"{'='*60}\n")

    # Initialize state
    initial_state: ResearchState = {
        "topic": topic,
        "search_queries": [],
        "search_results": [],
        "scraped_content": [],
        "analysis": "",
        "iteration": 0,
        "max_iterations": max_iterations,
        "needs_more_info": True,
        "final_report": "",
        "citations": []
    }

    # Create and run graph
    graph = create_research_graph()
    final_state = graph.invoke(initial_state)

    return final_state["final_report"]

````
